# DATA ANALYSIS OF IPL AUCTION (2013-23)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:\\Users\\ASUS\\Desktop\\ipl_data.csv')
print(df)
```

# Q - MOST EXPENSIVE PLAYER IN EACH YEAR 

```python

# Get most expensive player per season
top = df.loc[df.groupby("Season")["Price"].idxmax()].sort_values("Season")

# Plot
plt.figure(figsize=(12, 5))
ax = sns.barplot(data=top, x="Season", y="Price", hue="Season",palette="tab10",legend=False)

# Add player name + price on each bar
for i, row in enumerate(top.itertuples()):
    ax.text(i, row.Price + 0.1, f"{row.Name}\n₹{row.Price:.1f}",
            ha="center", fontsize=8, fontweight="bold")

plt.title("Most Expensive Player per IPL Season")
plt.xlabel("Season")
plt.ylabel("Price (₹ Crore)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
